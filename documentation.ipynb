### Scoring, Feedback and Level Progression
##### During the game, the user will be sent back to the last checkpoint, each checkpoint having a corresponding number. When the game starts a timer is displayed at the top left corner of the user's viewing screen. The program can only end when the user has completed/touched all the checkpoints. Completion of the game will result in a message and after clicking the message, the program will loop back to the splash screen. To pause and exit the game there will be a pause button at the top right corner of the user's viewing screen. After clicking pause, the user is given two options: Continue or Quit. By triggering Quit, the user is sent back to the splash screen and by clicking Continue, the user resumes the game. No progress or score is saved whatsoever and the no feedback given to the user if Quit is triggered.
