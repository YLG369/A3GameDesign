### *User Interaction*
##### The users will interact with my project through mouse clicks in the splash screen and keyboard commands while playing in the levels. Left click will select and activate the trigger boxes - similar to websites. Left/right keys and AD keys move the user while the W key, up key or space bar will cause the user to jump.

[Task: Identify and describe the core actions or operations of your project.]: #
[Instructions: List the main actions or functions the system will perform in response to user inputs or events. ### *Core Gameplay*
##### In the splashscreen, clicking on the play key will start a new game for the user and the quit key will exit the user from the game. In the game, the A/left key will make the user go left, the D/right key will make the user go right and the W/up key/space bar will make the user jump. If the user touches the spikes, it will take the user to the splash screen. If the user touches a checkpoint it will save the position and it will take the user back to the checkpoint. Completion of the game will take the user back to the splashscreen.

[Task: Describe how the system will give feedback to the user or track progress.]: #
[Instructions: Specify if there will be any form of scoring system, progress tracking, or feedback (e.g., score updates, messages for correct/incorrect actions). Indicate what triggers these feedback responses (e.g., completing a task or solving a challenge).]: #
### *Scoring, Feedback and Level Progression*
##### During the game, the user will be sent back to the last checkpoint, each checkpoint having a corresponding number. When the game starts a timer is displayed at the top left corner of the user's viewing screen. The program can only end when the user has completed/touched all the checkpoints. Completion of the game will result in a message and after clicking the message, the program will loop back to the splash screen. To pause and exit the game there will be a pause key at the top right corner of the user's viewing screen. After clicking pause, the user is given two options: Continue or Quit. By triggering Quit, the user is sent back to the splash screen and by clicking Continue, the user resumes the game. No progress or score is saved whatsoever and the no feedback given to the user if Quit is triggered.
